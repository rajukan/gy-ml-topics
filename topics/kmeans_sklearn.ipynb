```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=150, n_features=2,centers=3,cluster_std=.5,shuffle=True,random_state=0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='white', marker='o', edgecolors='black', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, init='random', n_init=10,max_iter=300,tol=1e-04, random_state=0)
y_km = km.fit_predict(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[y_km == 0,0],
            X[y_km == 0,1],
            s=50,c='lightgreen',marker='s', label='Cluster 1')
plt.scatter(X[y_km == 1,0],
            X[y_km == 1,1],
            s=50,c='blue', marker='v',label='Cluster 2')
plt.scatter(X[y_km == 2,0],
            X[y_km == 2,1],
            s=50,c='red',marker='o', label='Cluster 3')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s=50,c='yellow',marker='*', label='Centroids')
plt.legend(scatterpoints=1)
```
